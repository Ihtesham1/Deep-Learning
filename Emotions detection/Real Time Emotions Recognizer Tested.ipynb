```python
#from keras.models import load_model
from time import sleep
#from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
from tensorflow.keras.models import model_from_json
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
import time
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model

#model= model_from_json(open("fer.json","r").read())
face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
#model.load_weights('fer2.h5')
model = load_model("my_h5_model.h5")


cap=cv2.VideoCapture(0)
happy_count = 0
sad_count = 0
neutral_count = 0
surprise_count = 0
fear_count = 0
angry_count = 0
disgust_count = 0
while True:
    ret,test_img=cap.read()# captures frame and returns boolean value and captured image
    if not ret:
        continue
    gray_img= cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)

    faces_detected = face_classifier.detectMultiScale(gray_img, 1.32, 5)


    for (x,y,w,h) in faces_detected:
        #cv2.rectangle(test_img,(x,y),(x+w,y+h),(255,0,0),thickness=7)
        roi_gray=gray_img[y:y+w,x:x+h]#cropping region of interest i.e. face area from  image
        roi_gray=cv2.resize(roi_gray,(48,48))
        img_pixels = image.img_to_array(roi_gray)
        img_pixels = np.expand_dims(img_pixels, axis = 0)
        img_pixels /= 255

        predictions = model.predict(img_pixels)

        #find max indexed array
        max_index = np.argmax(predictions[0])

        emotions = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')
        predicted_emotion = emotions[max_index]
        if predicted_emotion == "neutral":
            millis = int(round(time.time()))
            name = 'F:/images/neutral/neutral' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            neutral_count = neutral_count + 1
            

                # To stop duplicate images
            #currentFrame = 1
        elif predicted_emotion == "disgust":
            
            millis = int(round(time.time()))
            name = 'F:/images/disgust/disgust' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            disgust_count = disgust_count + 1

                # To stop duplicate images
            #currentFrame += 1
            
        elif predicted_emotion == "happy":
            
            millis = int(round(time.time()))
            name = 'F:/images/happy/happy' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            happy_count = happy_count + 1

                # To stop duplicate images
            #currentFrame += 1
        elif predicted_emotion == "fear":
            
            millis = int(round(time.time()))
            name = 'F:/images/fear/fear' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            fear_count = fear_count + 1

                # To stop duplicate images
            #currentFrame += 1
            
        elif predicted_emotion == "sad":
            
            millis = int(round(time.time()))
            name = 'F:/images/sad/sad' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            sad_count = sad_count + 1

                # To stop duplicate images
            #currentFrame += 1
            
        elif predicted_emotion == "surprise":
            
            millis = int(round(time.time()))
            name = 'F:/images/surprise/surprise' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            surprise_count = surprise_count + 1

                # To stop duplicate images
            #currentFrame += 1
            
        elif predicted_emotion == "angry":
            
            millis = int(round(time.time()))
            name = 'F:/images/angry/angry' + str(millis) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, test_img)
            angry_count = angry_count + 1

                # To stop duplicate images
            #currentFrame += 1
            
                

        cv2.putText(test_img, predicted_emotion, (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
        
            
    resized_img = cv2.resize(test_img, (300, 300))
    cv2.imshow('Facial emotion analysis ',resized_img)
    
            



    if cv2.waitKey(10) == ord('q'):#wait until 'q' key is pressed
        break
        
        
        
total = happy_count+sad_count+neutral_count+surprise_count+fear_count+angry_count+disgust_count
happy=happy_count/total
sad=sad_count/total
fear=fear_count/total
surprise=surprise_count/total
angry=angry_count/total
disgust=disgust_count/total
neutral=neutral_count/total
print("Happy {0:.2f}%".format(happy))
print("Sad {0:.2f}%".format(sad))
print("Fear {0:.2f}%".format(fear))
print("Surprise {0:.2f}%".format(surprise))
print("Neutral {0:.2f}%".format(neutral))
print("Disgust {0:.2f}%".format(disgust))
print("Angry {0:.2f}%".format(angry))


cap.release()
cv2.destroyAllWindows








```
